```python
import pandas as pd
file_name=list(uploaded.keys())[0]
df = pd.read_csv (file_name)
df.head()
```

```python
#inspacting our data : we are understanding our data set

df.shape # check the number of rows and columns
df.info() # check column names ,types,and missing values
df.describe() # summery stats for numeric columns
df.isnull().sum() #count the missing values in each column
df.columns
```

```python
#question : what are the leading causes of death in the united state overall
#assumptions
df.head()
```

```python
#seleting relevant columns
df_relevant=df [['Cause Name','Deaths']]
```

```python
# total death per cause
deaths_by_cause=df_relevant.groupby("Cause Name")["Deaths"].sum()
```

```python
deaths_by_cause
```

```python
#sort causes by n of deaths
deaths_by_cause.sort_values(ascending=False)
```

```python
top_3_causes=deaths_by_cause.sort_values(ascending=False).head(3)
top_3_causes
```

```python
import matplotlib.pyplot as plt
```

```python
#visualisation :
import matplotlib.pyplot as plt
plt.figure() #creat a new figure ()
deaths_by_cause.plot(kind="barh") # create a bar chart ()
plt.title ("total death per cause")
plt.xlabel("Cause Name ")
plt.ylabel("Deaths")
plt.show()
```
